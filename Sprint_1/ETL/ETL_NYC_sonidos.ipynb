### se eliminan todas las filas donde '1_engine_presence' tiene valor -1 porque esto significa que no fue etiquetada en la anotacion

```python
# Eliminar filas donde '1_engine_presence' tiene valor -1
df_NYC_sonidos = df_NYC_sonidos[df_NYC_sonidos['1_engine_presence'] != -1]
```

### tratamos de ver el contenido de las columnas 'year', 'week' , 'day', 'hour'

```python
unique_years = df_NYC_sonidos['year'].unique()
unique_weeks = df_NYC_sonidos['week'].unique()
unique_days = df_NYC_sonidos['day'].unique()
unique_hours = df_NYC_sonidos['hour'].unique()

print("Años únicos:", unique_years)
print("Semanas únicas:", unique_weeks)
print("Días únicos:", unique_days)
print("Horas únicas:", unique_hours)
```

### se crea la columna fecha a partir de las columnas 'year'	'week'	'day'	'hour'

```python
#  'week' y 'day' son el número de la semana y el día de la semana respectivamente
# Convertir 'week' y 'day' a la fecha del primer día de la semana
df_NYC_sonidos['fecha'] = pd.to_datetime(df_NYC_sonidos['year'].astype(str) + 
                                         df_NYC_sonidos['week'].astype(str) + 
                                         df_NYC_sonidos['day'].astype(str), format='%Y%W%w')

# Sumar las horas correspondientes
df_NYC_sonidos['fecha'] += pd.to_timedelta(df_NYC_sonidos['hour'], unit='h')

df_NYC_sonidos
```

### losto todos los elementos dentro de la columna 'borough'

```python
# Obtener los elementos únicos de la columna 'borough'
boroughs_unicos = df_NYC_sonidos['borough'].unique()

# Imprimir los elementos únicos
for borough in boroughs_unicos:
    print(borough)
```

### transformamos los valores que aparecen en 'borough' , de tal manera que:
- 1=Manhattan
- 3=Brooklyn
- 4=Queens

```python
# Diccionario de mapeo de valores de 'borough'
mapeo_boroughs = {
    1: 'Manhattan',
    3: 'Brooklyn',
    4: 'Queens'
}

# Aplicar la transformación a la columna 'borough'
df_NYC_sonidos['borough'] = df_NYC_sonidos['borough'].map(mapeo_boroughs)

# Verificar el resultado
print(df_NYC_sonidos['borough'])
```

### se eliminan las columnas 'year'	'week'	'day'	'hour', una vez creada la columna fecha

```python
# Lista de columnas a eliminar
columnas_a_eliminar = [
    'year',
    'week',
    'day',
    'hour',
    'annotator_id',
    'block',
    'latitude',
    'longitude'
]

# Eliminar las columnas
df_NYC_sonidos = df_NYC_sonidos.drop(columnas_a_eliminar, axis=1)
```

```python
# Obtiene información sobre el DataFrame_victimas
print(df_NYC_sonidos.info())
```

cambiar nombre a las columnas 

```python
column_mapping = {
    'borough':'Barrio',
    '1-1_small-sounding-engine_presence': 'small_sounding_engine_presence',
    '1-2_medium-sounding-engine_presence': 'medium_sounding_engine_presence',
    '1-3_large-sounding-engine_presence': 'large_sounding_engine_presence',
    '1-X_engine-of-uncertain-size_presence': 'engine_of_uncertain_size_presence',
    '1_engine_presence': 'engine_presence'
}

# Renombrar las columnas utilizando el diccionario
df_NYC_sonidos = df_NYC_sonidos.rename(columns=column_mapping)
```

### ordenamos los valores del dataframe por fecha

```python
# Ordenar el DataFrame en función de 'start_date' en orden ascendente
df_NYC_sonidos = df_NYC_sonidos.sort_values(by='fecha', ascending=True)

# Verificar el resultado
print(df_NYC_sonidos.head())  # Muestra las primeras filas del DataFrame ordenado
```

### eliminamos las horas, minutos y segundos de las mediciones 

```python
# Extraer solo la parte de la fecha (año, mes, día)
df_NYC_sonidos['fecha'] = df_NYC_sonidos['fecha'].dt.date
```
