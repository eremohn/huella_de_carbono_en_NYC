```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv("Countries_CO2_Emission.csv")

# Filtra los datos para el pas "United States" y excluye 'all_energy_types'
us_data = df[(df['pais'] == 'United States') & (df['tipo_energia'] != 'all_energy_types')]

# Agrupa los datos por tipo de energa
grouped_data = us_data.groupby('tipo_energia')

# Configura el tamao del grfico
plt.figure(figsize=(10, 6))

# Itera sobre los grupos y grafica cada uno
for name, group in grouped_data:
    plt.plot(group['anio'], group['consumo_de_energia'], label=name)

# Aade etiquetas y ttulo
plt.xlabel('Ao')
plt.ylabel('Consumo de Energa')
plt.title('Consumo de Energa en Estados Unidos por Tipo de Energa')

# Aade una leyenda
plt.legend()

# Muestra el grfico
plt.show()
```
